# Correlation

Does the number of genre types released during a year have a relationship with each other?

# tv_kids1 Dataset

```python
tv_kids1['release_year'].corr(tv_kids1['British'])
```

```python
tv_kids1['release_year'].corr(tv_kids1['Spanish'])
```

The correlation between release year and British is negitive, meaning that as the years increase the British TV Shows 
release does not increase. 
The correlation between release year and Spanish genres is positive and as the years increase it is very slim that 
Spanish genres release would increase.

```python
# Dropping non-continuous variables

tv_kids12 = tv_kids1.drop(['show_id', 'release_year', 'Kids_TV'], axis=1)
tv_kids12.head(1)
```

```python
tv_kids12.corr(method='pearson')
```

```python
tv_kids12.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('cubehelix'), axis=1)
```
