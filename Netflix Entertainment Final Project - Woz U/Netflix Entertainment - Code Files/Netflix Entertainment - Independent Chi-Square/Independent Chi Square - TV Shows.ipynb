# Loading Datasets

```python
tv_kids = pd.read_csv("kids.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
tv_kids.head(1)
```

```python
tv_kids1 = tv_kids.drop(['title', 'listed_in', 'StreamR', 'date_added', 'country'], axis=1)
tv_kids1.dropna()
tv_kids1.head(1)  
```

```python
tv_doc = pd.read_csv("doc.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
tv_doc.head(1)
```

# Independent Chi Square Test Assumption Analysis

```python
# Contingency Table
british_crosstab = pd.crosstab(tv_kids1['release_year'], tv_kids1['British'])
british_crosstab
```

```python
# Running Independent Chi-Square
stats.chi2_contingency(british_crosstab)
```

British Genre: Looking at this, it looks like there is significant relationship between release year and British genre. 

```python
# Spanish Contingency Table 
spanish_crosstab = pd.crosstab(tv_kids1['release_year'], tv_kids1['Spanish'])
spanish_crosstab
```

```python
# Running Independent Chi-Square
stats.chi2_contingency(spanish_crosstab)
```

Spanish Genre: Looking at this, it looks like there is not significant relationship between release year and Spanish genre.
