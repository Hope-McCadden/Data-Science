# Importing Packages

```python
import pandas as pd
import numpy as np
from scipy.stats import norm
from scipy import stats
from scipy.stats import boxcox
import seaborn as sns
from scipy.stats import ttest_ind
```

# Loading datasets

```python
movieS1 = pd.read_csv("movieS1.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
movieS1.head(1)
```

```python
movieS12 = movieS1.drop(['title', 'listed_in', 'StreamR', 'date_added', 'country'], axis=1)
movieS12.dropna()
movieS12.head(1)  
```

```python
movieS2 = pd.read_csv("movieS2.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
movieS2.head(1)
```

```python
movieS22 = movieS2.drop(['title', 'listed_in', 'StreamR', 'date_added', 'country'], axis=1)
movieS22.dropna()
movieS22.head(1) 
```

```python
movieS3 = pd.read_csv("movieS3.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
movieS3.head(1)
```

```python
movieS32 = movieS3.drop(['title', 'listed_in', 'StreamR', 'date_added', 'country'], axis=1)
movieS32.dropna()
movieS32.head(1) 
```

# Subsetting Datasets

```python
mFamily1 = movieS12[['release_year', 'Documentaries', 'Independent']]
mFamily1.head(1)
```

```python
mFamily2 = movieS22[['release_year', 'Documentaries', 'Independent']]
mFamily2.head(1)
```

```python
mFamily3 = movieS32[['release_year', 'Documentaries', 'Independent']]
mFamily3.head(1)
```

# Concat datasets

```python
familyMovies = pd.concat([mFamily1, mFamily2, mFamily3])
familyMovies
```

# Movies: Independent Chi Square Test Assumption Analysis

Question: Are Documentaries films considered to be Independent films?

```python
# Contingency Table
movies_crosstab = pd.crosstab(familyMovies['Documentaries'], familyMovies['Independent'])
movies_crosstab
```

```python
stats.chi2_contingency(movies_crosstab)
```

Looking at this, it looks like there is significant relationship between Documentaries and Independent genres. 
Documentaries could be an independent film.
