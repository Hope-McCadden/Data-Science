```python
# importing packages 'pandas'

import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# import 'Netflix Entertainment Dataset'

netflix1 = pd.read_csv("netflix1.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
netflix1.head(1)
```

# Beginning of Data Wrangling

```python
# Value_counts() function to sum total TV_Shows and Movies

netflix1.StreamR.value_counts()
netflix1.StreamR.value_counts()
```

```python
# Remove 'Movie' rows from StreamR column

# Get names of indexes for which column Age has value 30
index = netflix1[ netflix1['StreamR'] == "Movie" ].index

# Delete these row indexes from dataFrame
netflix1.drop(index , inplace=True)
netflix1.head(1)
```

```python
# Split and rename Columns

netflix_TV = netflix1['listed_in'].str.split(',', expand=True).rename(columns = lambda x: 'Gengre'+str(x+1))
netflix_TV.rename(columns = {'Gengre1' : 'Kids_TV', 'Gengre2' : 'Crime', 'Gengre3' : 'Docuseries'}, inplace = True)
netflix_TV.head(1)
```

```python
nt = netflix_TV.drop(['Crime','Docuseries'], axis=1)
nt.head(1)
```

```python
# dummy coding kids_TV column

kidsDummy = pd.get_dummies(nt['Kids_TV'],drop_first=True)

kids1 = pd.concat([netflix1, nt, kidsDummy], axis = 1)
kids1.head(1)
```

```python
# Exporting dataframe: netflix to an csv file.
kids1.to_csv("kids1.csv", index = False) #Python
```

```python
kids1 = pd.read_csv("kids1.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
kids1.head(1)
```

# Beginning of Crime column 

```python
nt1 = netflix_TV.drop(['Kids_TV','Docuseries'], axis=1)
nt1.head(1)
```

```python
# dummy coding crime column
crimeDummy = pd.get_dummies(nt1['Crime'], drop_first=True)

# Concat 2 datasets 'tv' and 'crimeDummy'
crime = pd.concat([netflix1, nt1, crimeDummy], axis=1)
crime.head(1)
```

```python
crime = pd.read_csv("crime.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
crime.head(1)
```

# Beginning of Docuseries column

```python
nt2 = netflix_TV.drop(['Kids_TV','Crime'], axis=1)
nt2.head(1)
```

```python
# dummy coding Docuseries column
docDummy = pd.get_dummies(nt2['Docuseries'],drop_first=True)

doc = pd.concat([netflix1, nt2, docDummy], axis=1)
doc.head(1)
```

```python
# Exporting dataframe: crime to an csv file.
doc.to_csv("doc.csv", index = False) #Python
```

```python
doc = pd.read_csv("doc.csv")
pd.options.display.max_rows = 10000
pd.options.display.max_columns = 10000
doc.head(1)
```
